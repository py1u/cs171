```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
teams = pd.read_csv("teams.csv")
teams
```

```python
X = teams[["athletes", "prev_medals"]].copy()
y = teams["medals"].copy()
```

```python
X
```

```python
y
```

```python
X["intercept"] = 1
X = X[["intercept", "athletes", "prev_medals"]]
X
```

```python
X_T = X.T
X_T
```

```python
B =  np.linalg.inv(X_T @ X ) @ X_T @ y 
B
```

```python
B.index = X.columns
B
```

```python
predictions = X @ B
predictions
```

```python
SSR =((y - predictions) ** 2).sum()
```

```python
SST = ((y - y.mean()) ** 2).sum()
```

```python
SSR
```

```python
SST
```

```python
R2 = 1 - (SSR/SST)
R2
```
